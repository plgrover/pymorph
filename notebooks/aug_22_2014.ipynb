## First Order Morphodynamic Model

```python
x = []
zb = []
with open(r'C:\Users\Patrick\Dropbox\PhD\Morphodynamic_Model\data\BedProfile_eq_v03.csv','rb') as csvfile:
    reader = csv.reader(csvfile,delimiter=',')
    for row in reader:
        x.append(float(row[0])/100.)
        zb.append(float(row[1])/100.)
dx = 0.03

zb0 = copy.deepcopy(zb)
number_nodes = len(x)
z = np.ones(number_nodes) * 0.3
qb = np.zeros(number_nodes)
qf = 0.08
A = 1.0
u = qf/(z-zb)
qbs = A * u
timesteps = 550
dt = 0.01

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,zb0, label='bed at t=0')
for t in range(timesteps):
    zbn = copy.deepcopy(zb)
    for i in range(1,number_nodes):
        ''' Using a simple first order upwind difference equation '''
        zb[i] = zbn[i] - (dt/(dx))*(qbs[i]-qbs[i-1])
    
    ''' Update the boundary conditions '''
    #zb[0]= zb[number_nodes-1] 
    zb[0] = zbn[0] - (dt/(x[i]-x[i-1]))*(qbs[i]-qbs[i-1])
    ''' Update the velocity and bedload '''
    h = z - zb
    u = qf/h
    qbs = A * u
    
ax1.plot(x,zb, label='bed at t=1s')

''' Now calculate the area under the curve, both at t=0 and the final timestep'''
area_t0 = simps(np.asarray(zb0),dx=dx)
area_tf = simps(np.asarray(zb),dx=dx)

print 'Initial area %s, final area %s'%(area_t0,area_tf)

plt.show()
```

It looks like we are actually gaining volume over time which is not good.

## Second-order upwind scheme

A second order upwind scheme is applied below. Note that I changed the inital bed profile so that the $\Delta x$ is consistent allowing use of a fintie differences approach.

```python
x = []
zb = []
with open(r'C:\Users\Patrick\Dropbox\PhD\Morphodynamic_Model\data\BedProfile_eq_v03.csv','rb') as csvfile:
    reader = csv.reader(csvfile,delimiter=',')
    for row in reader:
        x.append(float(row[0])/100.)
        zb.append(float(row[1])/100.)

dx = 0.03

zb0 = copy.deepcopy(zb)
number_nodes = len(x)
z = np.ones(number_nodes) * 0.3
qb = np.zeros(number_nodes)
qf = 0.08
A = 1.0
u = qf/(z-zb)
qbs = A * u
timesteps = 30
dt = 0.01

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,zb0, label='bed at t=0')
for t in range(timesteps):
    zbn = copy.deepcopy(zb)
    for i in range(2,number_nodes-1):
        ''' Using a  second order upwind difference equation '''
        zb[i] = zbn[i] - (dt/(6. * dx))*(2.*qbs[i+1] + 3.*qbs[i] - 6.*qbs[i-1] + qbs[i-2])
    
    ''' Update the boundary and near-boundary conditions '''
    ''' i = 1'''
    zb[1] = zbn[1] - (dt/(6. * dx))*(2.*qbs[2] + 3.*qbs[1] - 6.*qbs[0] + qbs[number_nodes-1])
    ''' i = 0'''
    zb[0] = zbn[0] - (dt/(6. * dx))*(2.*qbs[1] + 3.*qbs[0] - 6.*qbs[number_nodes-1] + qbs[number_nodes-2])    
    ''' i = number_nodes -1 '''
    zb[number_nodes-1] = zbn[number_nodes-1] - (dt/(6. * dx))*(2.*qbs[0] + 3.*qbs[number_nodes-1] - 6.*qbs[number_nodes-2] + qbs[number_nodes-3])
    
    ''' Update the velocity and bedload '''
    h = z - zb
    u = qf/h
    qbs = A * u
    
ax1.plot(x,zb, label='bed at t=1s')

''' plot the gradient in the bedload '''
grdqb = np.zeros(number_nodes)
for i in range(2,number_nodes-1):
    ''' Using a  second order upwind difference equation '''
    grdqb[i] = (1./(6. * dx))*(2.*qbs[i+1] + 3.*qbs[i] - 6.*qbs[i-1] + qbs[i-2])

ax2 = ax1.twinx()
ax2.plot(x,grdqb, label='grad qbs')

plt.show()


```

## August 22,2014 Increase $\Delta x$

Going to investigate what happens if we increase the spatial resolution of the data. Also changed the time-marching to MacCormack's method:
$$\tilde{u}_{n+1} = u_n + \Delta tu_n'$$
$$ u_{n+1} = u_n + \frac{1}{2}\Delta t\left(u_n' + \tilde{u}_{n+1}' \right)$$


```python
''' Written as a function '''
def morphmodel_2ndO_upwind(num_cells):
    from scipy.interpolate import interp1d
    
    x=[]
    zb=[]
    with open(r'C:\Users\Patrick\Dropbox\PhD\Morphodynamic_Model\data\BedProfile_eq_v03.csv','rb') as csvfile:
        reader = csv.reader(csvfile,delimiter=',')
        for row in reader:
            x.append(float(row[0])/100.)
            zb.append(float(row[1])/100.)
    
    xnew = np.linspace(min(x),max(x), num=num_cells)
    f1 = interp1d(x,zb)
    zbnew = f1(xnew)
    
    ''' Copy the variables '''
    x = copy.deepcopy(xnew)
    zb = copy.deepcopy(zbnew)
    zb0 = copy.deepcopy(zb)
    
    dx = x[1] - x[0]
    
    number_nodes = len(x)
    z = np.ones(number_nodes) * 1.3
    qb = np.zeros(number_nodes)
    grdqb = np.zeros((number_nodes,2))
    
    qf = 0.45
    A = 1.0
    u = qf/(z-zb)
    qbs = A * u
    timesteps = 8000
    dt = 0.005
    
    fig = plt.figure(figsize=(16, 4))
    ax1 = fig.add_subplot(111)
    ax1.plot(x,zb0, label='bed at t=0')
    
    plotupdate = 0
    
    for t in range(timesteps):
        zbn = copy.deepcopy(zb)    
        for stage in range(2):        
            ''' ----------------------------- '''
            ''' Calculate the derivative      '''
            ''' ----------------------------- '''
            for i in range(2,number_nodes-1):
                ''' Using a  second order upwind difference equation '''
                grdqb[i,stage] = (1./(6. * dx))*(2.*qbs[i+1] + 3.*qbs[i] - 6.*qbs[i-1] + qbs[i-2])
                
                if abs(grdqb[i,stage]) > 0.01:
                    grdqb[i,stage] = (1./(dx))*(qbs[i]-qbs[i-1])
                
            ''' Update the boundary and near-boundary conditions '''
            ''' i = 1'''
            grdqb[1,stage] = (1./(6. * dx))*(2.*qbs[2] + 3.*qbs[1] - 6.*qbs[0] + qbs[number_nodes-1])
            ''' i = 0'''
            grdqb[0,stage] = (1./(6. * dx))*(2.*qbs[1] + 3.*qbs[0] - 6.*qbs[number_nodes-1] + qbs[number_nodes-2])        
            ''' i = number_nodes -1 '''
            grdqb[number_nodes-1,stage] = (1./(6. * dx))*(2.*qbs[0] + 3.*qbs[number_nodes-1] - 6.*qbs[number_nodes-2] + qbs[number_nodes-3])
            
            for i in range(number_nodes):
                if stage == 1:
                    zb[i] = zbn[i] - dt*grdqb[i,0] 
                else:
                    zb[i] = zbn[i] - 0.5*dt*(grdqb[i,0] +  grdqb[i,1])
        
            ''' Update the velocity and bedload '''
            h = z - zb
            u = qf/h
            qbs = A * u
        
        if plotupdate == 2000:
            tim = float(t)*dt
            ax1.plot(x,zb,label='t=%s' % tim )
            plotupdate = 0
        else:
            plotupdate +=1
            
    tim = float(timesteps)*dt
    ax1.plot(x,zb, label='bed at t=%s' % tim)
    #ax1.plot(x0,zb_1st,'r-', label='bed at t=1s')
    
    ''' Now calculate the area under the curve, both at t=0 and the final timestep'''
    area_t0 = simps(np.asarray(zb0),dx=dx)
    area_tf = simps(np.asarray(zb),dx=dx)
    
    print 'Initial area %s, final area %s'%(area_t0,area_tf)
    print 'Mean velocity: %s' % mean(u)
    plt.legend()
    plt.savefig('morphmodel_2ndO_upwind_%s.png' % num_cells)
    
    return x,zb,grdqb[:,1]
```

```python
''' GRID sensitivity Test '''
from IPython.display import Image

x101,z101,grdqb101 = morphmodel_2ndO_upwind(101)
x501,z501,grdqb501 = morphmodel_2ndO_upwind(501)
x1001,z1001,grdqb1001 = morphmodel_2ndO_upwind(1001)

Image('morphmodel_2ndO_upwind_101.png')
Image('morphmodel_2ndO_upwind_501.png')
Image('morphmodel_2ndO_upwind_1001.png')
```

```python
fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x101,z101, label='101 cells')
ax1.plot(x501,z501, label='501 cells')
ax1.plot(x1001,z1001, label='1001 cells')
plt.legend()
plt.show()
```

What we see is that the model appears to be grid insensitive between 501-1001 cells. Interestingly, as we increase the number of cells we also increase the ability of the model to capture the front of the wave with less dissipation. Clearly for 101 cells the dissipation is significant. The wave travel time is also impacted by the grid resolution as well - however this may be tied to the dissipation of the crest - the shorter the crest, the slower the wave will move under this experiment. 

The model uses some artifical diffusion, basically switching to a first-order upwind method when the gradient of the sediment transport is greater than 0.01. 

```python
fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x101,grdqb101,label='101')
ax1.plot(x501,grdqb501,label='501')
ax1.plot(x1001,grdqb1001,label='1001')

artdiff = np.ones(1001) * 0.01
ax1.plot(x1001,artdiff,'--')
artdiff = np.ones(1001) * - 0.01
ax1.plot(x1001,artdiff,'--')
plt.legend()
plt.show()
```
