## Dune Geometry

First will develop the model geometry. Will use the actual bed geometry from the lab. 

```python
import csv
import copy
x = []
zb = []
with open(r'C:\Users\Patrick\Dropbox\PhD\Morphodynamic_Model\data\BedProfile_v02.csv','rb') as csvfile:
    reader = csv.reader(csvfile,delimiter=',')
    for row in reader:
        x.append(float(row[0])/100.)
        zb.append(float(row[1])/100.)
        
zb0 = copy.deepcopy(zb)

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,zb,'k.', label='bed')
plt.xlabel('x (m)')
plt.ylabel('Elevation (m)')
plt.show()
```

Now set up the flow depth and other variables.

```python
number_nodes = len(x)
z = np.ones(number_nodes) * 0.3
qb = np.zeros(number_nodes)
```

For simplicity we will continue to assume that the depth-averaged velocity, $u$, is a function of the flow depth, $h$, and the inflow, $q_f$. 

$$u=\frac{q_f}{h}$$

ad we will also assume that the equilibrium sediment transport, $q_b^*$, is a function of the depth averaged velocity:

$$q_b^*=Au$$

The flow parameters are shown here:

```python
qf = 0.08
A = 1.0
u = qf/(z-zb)
qbs = A * u
print mean(u)
```

```python
fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,qbs, label='bedload')
plt.xlabel('x (m)')
plt.ylabel('Equilibrium Bedload (m2/s)')
plt.show()
```

No surprisingly the bedload looks much like the profile of the bed. For the intial try, will apply the equilibrium bedload function. Note that I have applied a basic fintied differnece approach however, the node spacing, $\Delta x$ is not constant. Need to apply some form of fintie volume approach. Also looks like we are loosing sediment. 

```python
timesteps = 550
dt = 0.01

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,zb0, label='bed at t=0')
for t in range(timesteps):
    zbn = deepcopy(zb)
    for i in range(1,number_nodes):
        ''' Using a simple first order upwind difference equation '''
        zb[i] = zbn[i] - (dt/(x[i]-x[i-1]))*(qbs[i]-qbs[i-1])
    
    ''' Update the boundary conditions '''
    zb[0]= zb[number_nodes-1] #zbn[0] - (dt/(x[1]-x[0]))*(qbs[0]-qbs[number_nodes-1])
    #zb[number_nodes-1] = zb[number_nodes-2]
    
    ''' Update the velocity and bedload '''
    h = z - zb
    u = qf/h
    qbs = A * u
    
ax1.plot(x,zb, label='bed at t=1s')

plt.show()
```

The change in the profile seems to show that the dunes are progressively flattening out.
