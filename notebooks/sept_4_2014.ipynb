First run the code without the slope modification term.

```python
x,zb = single_hump( 20., 201 )
zb0 = copy.deepcopy(zb)

depth = 2.5
qf = 0.45
timesteps = 1000
dt = 0.01
x,zb,u,qbs = morphmodel_2ndO_upwind(x,zb,depth,qf,timesteps,dt,False)

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,zb, label='bed')
ax1.plot(x,zb0, label='initial bed')
plt.show()
```

Now re-run using the slope modifier. Here we are using the slope modification from Equation 87 in Dey and Papanicolaou 2008..

$$ Slope Adjustment = cos(\theta)\left(1 - \frac{tan(\theta)}{tan(\phi)} \right) $$

```python
x,zb = single_hump( 20., 201 )
zb0 = copy.deepcopy(zb)

depth = 2.5
qf = 0.45
timesteps = 1000
dt = 0.01
x,zb,u,qbs = morphmodel_2ndO_upwind(x,zb,depth,qf,timesteps,dt, True)

fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(111)
ax1.plot(x,zb, label='bed')
ax1.plot(x,zb0, label='initial bed')
plt.show()
```
